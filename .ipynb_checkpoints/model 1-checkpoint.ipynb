```python
from ib_insync import *
util.startLoop()

ib = IB()
ib.connect('', 7497, clientId=4)
```

Grab some data. see docs . https://rawgit.com/erdewit/ib_insync/master/docs/html/

```python
ib.isConnected()
```

```python
contract = Stock('TSLA', 'SMART', 'USD')

ib.reqHeadTimeStamp(contract, whatToShow='TRADES', useRTH=True)
```

```python
# ib.positions()
```

```python
# TRADES HISTORICAL_VOLATILITY OPTION_IMPLIED_VOLATILITY
```

```python
bars = ib.reqHistoricalData(
        contract,
        endDateTime='',
        durationStr='90 D',
        barSizeSetting='1 min',
        whatToShow='TRADES',
        useRTH=True,
        formatDate=1)
```

```python
bars_histVol = ib.reqHistoricalData(
        contract,
        endDateTime='',
        durationStr='90 D',
        barSizeSetting='1 min',
        whatToShow='HISTORICAL_VOLATILITY',
        useRTH=True,
        formatDate=1)
```

```python
bars_impVol = ib.reqHistoricalData(
        contract,
        endDateTime='',
        durationStr='90 D',
        barSizeSetting='1 min',
        whatToShow='OPTION_IMPLIED_VOLATILITY',
        useRTH=True,
        formatDate=1)
```

```python
bars_histVol[0]
# bars2[0]
```

```python
# convert into dataframes 
df = util.df(bars)
df_histVol = util.df(bars_histVol)
df_impVol = util.df(bars_impVol)

display(df.shape)
display(df.head())
display(df.tail())

# display(df_histVol.shape)
# display(df_histVol.head())
# display(df_histVol.tail())

display(df_impVol.shape)
display(df_impVol.head())
display(df_impVol.tail())
```

```python
# the historic volatility data is sparse. Will just focus on trades and implied volitility.
# remove empty colums (volume, barcount) and rename columns so they dont conflict with Trades
df_impVol.drop(columns=['volume', 'barCount'], inplace=True)
```

```python
# Rename cols
df_impVol.rename(index=str, columns={"open": "volOpen", "high":"volHigh", "low":"volLow", "close":"volClose", "average":"volAverage"})
```

```python
# Rename cols
df_impVol.head()
```

```python
# Merge datasets
df_merged = df.merge(df_impVol, left_on='date', right_on='date', how='outer')
```

```python
# add new datapoint . 1 for direction (price up/down). 2 magnitude as % of 
# adding rolling mean for price 
import pandas as pd

df_merged.head(2)
df = df_merged

# df['rollPrice'] = pd.rolling_mean(df['close_x'], 200)
df['trend'] = df['open_x'] - df['close_x']
df.head(2)
```

```python
# create a score for rolling means 
df['rollPrice'] = df['close_x'].rolling(200).mean()
# df['trend'] = df['open_x'] - df['close_x']
df.head(205)
```
